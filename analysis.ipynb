```python
import pandas as pd

filepaths = '강원특별자치도 춘천시_의류수거함 현황_20240130.csv'

try:
    data = pd.read_csv(filepaths, encoding='utf-8')
except:
    data = pd.read_csv(filepaths, encoding='cp949')

data.head(94)
```

```python
output_filepaths = '강원특별자치도 춘천시_의류수거함 현황_변형된_데이터.csv'
data.to_csv(output_filepaths, index=False, encoding='utf-8-sig')
```

```python
#데이터 확인
data = pd.read_csv(output_filepaths, encoding='utf-8')
print(data.head())
```

```python
for column in data.columns:
    if data[column].dtype == 'float64' or data[column].dtype == 'int64':    #숫자형 데이터
        data[column].fillna(data[column].mean(), inplace=True)  #평균값으로 결측치 NaN(Not Available) 대체, 원본 데이터프레임 수정
    elif data[column].dtype == 'object':    #문자열 데이터
        data = data.dropna(subset = [column])   #결측치 제거


admin_district = data['읍면동'].value_counts()  #읍면동별 수거함 개수
print(admin_district)
```

```python
#읍면동별 인구 데이터
population_data= {
    '석사동': 33473, '효자동' : 21164, '퇴계동': 47164, '후평동': 44926, '동내면': 16809, 
    '옥천동': 8894, '신동면': 2545, '소양동': 8894, '남산면': 3270, '사북면': 2363,
    '동면': 19104, '서면' : 3609, '신동,사농동' :25607, '죽림동': 5429, '온의동': 25589, '조양동,운교동':2533
}

#읍면동별 수거함 데이터프레임 생성
admin_district_df = pd.DataFrame(admin_district).reset_index()
admin_district_df.columns = ['Distirct', 'Count']
admin_district_df['Population'] = admin_district_df['Distirct'].map(population_data)

#1만명당 수거함 개수 계산
admin_district_df['Count_per_capita'] = admin_district_df['Count'] / admin_district_df['Population'] * 10000

#설치 연도를 datetime 형식으로 변환
data['설치년월'] = pd.to_datetime(data['설치년월'], errors='coerce')
data['설치연도'] = data['설치년월'].dt.year

#읍면동별 평균 설치 연도 계산
average_year = data.groupby('읍면동')['설치연도'].mean().reset_index()
average_year.columns = ['Distirct', 'Average_year']

#최종 데이터 결합
bins_analysis_df=  admin_district_df.merge(average_year, on='Distirct', how='left')

#주요 대로변에 설치된 수거함 개수 확인
main_road = data[data['도로명주소'].str.contains('대로|로', na=False)]
main_roda_count = main_road['읍면동'].value_counts().reset_index()
main_roda_count.columns = ['Distirct', 'Main_road_count']
print("\n주요 대로변에 설치된 수거함 개수")
print(main_roda_count)
```
